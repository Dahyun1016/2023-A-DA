```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []

temp_day = []
temp_mean = []
year = []

for row in data:
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '':
            temp_day.append(float(row[2]))  # Fixed this line, added float conversion
            if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '31':
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day) / len(temp_day))  # Fixed this line, moved the closing parenthesis
                temp_day = []

f.close()

plt.figure(figsize=(10, 4))
plt.rc('font', family='Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year, temp_mean, color='k', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

aug_temperatures = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '' and month == '08':
        aug_temperatures.append(float(row[4]))

f.close()

plt.hist(aug_temperatures, bins=100, color='r')
plt.title('August Temperature Histogram')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []



for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))


plt.boxplot(jan)
plt.boxplot(aug)

plt.show()
```
