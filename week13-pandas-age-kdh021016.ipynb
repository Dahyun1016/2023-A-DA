```python
import numpy as np
import pandas as pd

index = pd.date_range('12/1/2024', periods=8)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))

df
```

```python
df['F'] = np.ones(8)
df.head(3) # 앞에부분 세개만 보여줌
```

```python
df1 = df.drop(['E','F'],axis='columns') # olumns=axis=1 drop=삭제
df.head()
```

```python
df1.columns=['a','b','c']
df1
```
